```python
# Python libraries
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from tqdm import trange

# Custom functions
import graphGym
# from graphGym import GraphBandit
```

```python
# Define grid graph
# 8x8 grid
# node (0, 0) is soure, node (7, 7) is target

G = nx.grid_2d_graph(8, 8)
source_node = list(G.nodes)[0]
target_node = list(G.nodes)[-1]

def gridPos(cols, rows, seed = 101010):
    np.random.seed(seed)
    pos = {}
    for i in range(rows):
        for j in range(cols):
            pos[(j, i)] = (j + np.random.random() * 0.1, 
                           i + np.random.random() * 0.1)
    return pos

fig, ax = plt.subplots(1, 1, figsize = (5, 5))
ax.axis("off")

node_color =  ["#f9d531"] + ["#62b6de"]*(8*8 - 2) + ["#ef562f"]
nx.draw_networkx(G, pos = gridPos(10, 8), ax = ax, node_color = node_color, labels = {},
                width = 2, edgecolors = "black", linewidths = 2)
```
