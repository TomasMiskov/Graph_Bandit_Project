```python
import pickle
import numpy as np
import os
import matplotlib.pyplot as plt
```

### Load all resutls

```python
def extractRegrets(pickled_results_path):
    # Load results
    with open(pickled_results_path, "rb") as file:
        results = pickle.load(file)
        
    # Gather all regrets
    regrets = np.zeros(shape = (len(results), 100))
    run_times = np.zeros(len(results))
    for i in range(len(results)):
        regrets[i] = results[i].expectedRegret()
        run_times[i] = results[i].total_time
    
    # Average regrets
    avg_regret = regrets.mean(axis = 0)
    std_regret = regrets.std(axis = 0)
    avg_time = run_times.mean()
    
    return avg_regret, std_regret, avg_time

avg_regrets = {}
std_regrets = {}
avg_run_times = {}
for result in os.listdir("./Results"):
    avg_regrets[result], std_regrets[result], avg_run_times[result] = extractRegrets(r"./Results/" + result)
```

### Plot expected regrets

```python
myStyle = r"C:\Users\misko\miniconda3\Lib\site-packages\matplotlib\mpl-data\stylelib\nicePlots.mplstyle"

to_plot = [True, False, False, True, False, False, True, True]
labels = ["Dijkstra's Agent", "Dijkstra's Agent (c = 0.2)", "Dijkstra's Agent (c = 20)", "Local Agent", 
          "Sampling Agent (k = 1)", "Sampling Agent (k = 3)", "Sampling Agent (k = 5)", "Static Agent"]
markers = ['o', 'v', '^', '*', 's', 'x', 'D', ',']

with plt.style.context(myStyle):
    fig, ax = plt.subplots(1, 1, figsize = (10, 6))
    
    i = 0
    for key, val in avg_regrets.items():
        if to_plot[i]:
            ax.plot(val.cumsum(), label = labels[i], lw = 1, linestyle = '-', marker = markers[i], markevery = 10)
            
            # Shaded 95% confidence bounds
            ax.fill_between(range(100), val.cumsum() + 1.96* std_regrets[key], 
                            val.cumsum() - 1.96*std_regrets[key], alpha = 0.5)
        
        i += 1
    
    ax.set_xlabel("Iterations", fontsize = 12)
    ax.set_ylabel("Expected Regret", fontsize = 12)
    ax.legend()
    
    plt.savefig(".\Plots\expected_regret.png")
```

### Graph Plots

```python
# PLOTTING
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import numpy as np
import helperFunctions as helpers
import networkx as nx

def gridPos(cols, rows, seed = 101010):
    np.random.seed(seed)
    pos = {}
    for i in range(rows):
        for j in range(cols):
            pos[(j, i)] = (j, i)
    return pos

def plot8x8Grid(G, ax = None, node_color = ["#f9d531"] + ["#62b6de"]*(8*8 - 2) + ["#ef562f"], 
                     save_fig = False, **kwargs):
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize = (5, 5))
    
    ax.axis("off")
    ax.set_xlim(-.5, 7.5)
    ax.set_ylim(-.5, 7.5)
    
    grid_size = int(len(G.nodes)**.5)
    nx.draw_networkx(G, pos = gridPos(grid_size, grid_size), ax = ax, node_color = node_color, labels = {},
                    width = 3, edgecolors = "black", linewidths = 0)

    if save_fig:
        plt.savefig("./Plots/grid.png")
        
def plotTraveresedPaths(G, paths, ax = None, node_color = ["#f9d531"] + ["#62b6de"]*(8*8 - 2) + ["#ef562f"], 
                        save_fig = False, **kwargs):
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize = (5, 5))
        
    ax.axis("off")
    ax.set_xlim(-.5, 7.5)
    ax.set_ylim(-.5, 7.5)
    
    grid_size = int(len(G.nodes)**.5)
    nx.draw_networkx_nodes(G, pos = gridPos(grid_size, grid_size), ax = ax, node_color = node_color)

    for i in range(len(paths)):
        nx.draw_networkx_edges(G, pos = gridPos(grid_size, grid_size), ax = ax, edgelist = paths[i], 
                               width = [-np.array(env.G.edges[e]["rewards"]).mean()/2 for e in env.G.edges], **kwargs)
        
def plotUniquePaths(G, df_paths, ax = None, node_color = ["#f9d531"] + ["#62b6de"]*(8*8 - 2) + ["#ef562f"], 
                    edge_vmin = -1, edge_vmax = 50, save_fig = False, title = None):
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize = (5, 5))

    ax.axis("off")
    ax.set_xlim(-.5, 7.5)
    ax.set_ylim(-.5, 7.5)

    grid_size = int(len(G.nodes)**.5)
    nx.draw_networkx_nodes(G, pos = gridPos(grid_size, grid_size), ax = ax, node_color = node_color)
    
    edge_width = [-np.array(G.edges[e]["rewards"]).mean()/2 for e in G.edges]
    for i, path in enumerate(df_paths.path_edges):
        nx.draw_networkx_edges(G, pos = gridPos(grid_size, grid_size), ax = ax, edgelist = path, 
                               edge_color = [df_paths.counter[i]] * len(path), width = edge_width, 
                               edge_cmap = cm.Greys, edge_vmin = edge_vmin, edge_vmax = edge_vmax)
        
    ax.set_title(title)
        
    if save_fig:
        plt.savefig(r"./Plots/" + f"{title.replace(' ', '_')}.png")
```

### Show the difference in estimated edge weights between algorithms

```python
#-------------------------------
# Get mean edge-weight estimates
#-------------------------------
example_graphs = [dijkstra_results[3].G, local_results[3].G, sampling_results[3].G, static_results[3].G]
G = example_graphs[0]

estimated_weights = np.zeros(shape = (5, 112))
estimated_weights[0] = np.array([-G.edges[e]["expected_reward"] for e in G.edges])
for i, G in enumerate(example_graphs):
    estimated_weights[i + 1] = np.array([-np.array(G.edges[e]["rewards"]).mean() for e in G.edges])
    
# Sort them according to ground truth
sort_indices = np.argsort(estimated_weights[0])

for i in range(len(estimated_weights)):
    estimated_weights[i] = estimated_weights[i][sort_indices]
    
#------
# Plot
#------
fig, axes = plt.subplots(5, 1, figsize = (12, 2.5), gridspec_kw = {'hspace': 1})
y_labels = ["Ground Truth", "Dijkstra's Agent", "Local Agent", "Sampling Agent", "Static Agent"]

for i in range(len(y_labels)):
    axes[i].imshow(estimated_weights[i].reshape(-1, 1).T, cmap = cm.Blues, aspect = "auto")
    axes[i].set_yticks([0], [y_labels[i]], font = "Lucida Sans")
    
    for key, spine in axes[i].spines.items():
        spine.set_visible(False)
    
    axes[i].get_xaxis().set_visible(False)
    axes[i].yaxis.set_ticks_position("none")
    
    if i == 0:
        axes[i].get_xaxis().set_visible(True)
        axes[i].set_xticks([])
        axes[i].set_xlabel("Ground Truth Weight-Sorted Graph Edges", fontsize = 12, 
                           labelpad = 15, font = "Lucida Sans")    
        axes[i].xaxis.set_label_position("top") 
                         
plt.savefig(r"./Plots/estimated_edge_weights.png")
```
